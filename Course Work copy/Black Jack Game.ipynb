```python
class Card():
    
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return (self.rank + " of " + self.suit)
```

```python
class Deck(Card):
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank))
                
    def shuffle(self):
        random.shuffle(self.deck)
        
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n'+ card.__str__()
        return "The deck has " + deck_comp
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
    
```

```python
def hitorstand(deck, hand):
    global playing #to control an upcoming while loop
    
    while True:
        x = input("Hit or stand? Enter 'h' to hit, and 's' to stand! ")
        
        if x[0].lower() == 'h':
            hit(deck,hand)
        elif x[0].lower() == 's':
            print("Player stands. Dealer's turn.")
            playing = False
        else:
            print("Sorry, please enter 'h' or 's' only. ")
            continue
            
        break
```
